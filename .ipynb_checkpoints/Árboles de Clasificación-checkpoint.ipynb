## Árboles de Clasificación

```python
## Importamos librerías
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.tree import export_graphviz
from sklearn import tree
```

```python
df = pd.read_csv('./Datasets/titanic_clean.csv')
df.head()
```

```python
X_data = df.drop(['Unnamed: 0', 'Survived'], axis = 1)
X_data.head()
```

```python
y_survived = df['Survived']
y_survived.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_data, y_survived, test_size = 0.3, random_state = 42)
```

```python
dtcModel = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
dtcModel.fit(X_train, y_train)
```

```python
y_predict = dtcModel.predict(X_test)
```

```python
cm = confusion_matrix(y_test, y_predict)
print(cm)
```

```python
sns.heatmap(cm, annot = True, fmt = 'g', cmap = 'Oranges')
plt.show()
```

```python
print(classification_report(y_test, y_predict))
```

```python
export_graphviz(dtcModel, out_file = 'treeArbol.dot', filled = True, feature_names = ['Pclass', 'Sex','SibSp','Parch', 'Fare', 'Embarked'])
```

```python
plt.figure(figsize = (60,60))
tree.plot_tree(dtcModel, filled = True, feature_names= ['Pclass', 'Sex','SibSp','Parch', 'Fare', 'Embarked'])
plt.show()
```
